```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/LearnDataSci/article-resources/master/Housing%20Market/IMDB-Movie-Data.csv'
df = pd.read_csv(url)
```

```python
print(df.head())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Runtime (Minutes)', y='Rating', data=df, hue='Metascore', palette='viridis')
plt.title('Movie Duration vs. Audience Rating')
plt.show()
```

```python
director_stats = df.groupby('Director')['Rating'].mean().sort_values(ascending=False).head(10)
print("\nTop 10 Directors by Average Rating:\n", director_stats)

sns.boxplot(x='Rating', y='Genre', data=df)
plt.title('Rating Distribution by Genre')
plt.show()
```

```python
import numpy as np
```

```python
df['Decade'] = (df['Year'] // 10) * 10
```

```python
decade_trends = df.groupby('Decade')['Rating'].mean()
decade_trends.plot(kind='line', marker='o', color='purple', title='Average Movie Ratings by Decade')
plt.show()
```

```python
Q1 = df['Votes'].quantile(0.25)
Q3 = df['Votes'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Votes'] < (Q1 - 1.5 * IQR)) | (df['Votes'] > (Q3 + 1.5 * IQR))]
```

```python
print(f"Detected {len(outliers)} outlier movies (based on vote count).")

plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Votes', y='Rating', hue='Genre', size='Metascore', sizes=(20, 200), alpha=0.6)
plt.title('Multivariate Analysis: Rating vs Votes (Size=Metascore, Color=Genre)')
plt.xscale('log')
plt.show()
```
